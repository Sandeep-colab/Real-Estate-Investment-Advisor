```python
# Cell 1: Setup and Data Loading
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
pd.set_option('display.float_format', lambda x: '%.2f' % x) # Format floats nicely

# Load the dataset (Adjust path if needed)
try:
    df = pd.read_csv('../data/india_housing_prices.csv')
    print("Data loaded successfully.")
except FileNotFoundError:
    print("Error: File not found. Check the path: ../data/india_housing_prices.csv")
    
# Display initial info (as you provided)
print("\nDataFrame Info:")
df.info()
```

```python
# Cell 2: Target Variable Engineering

# --- Regression Target ---
def create_future_price_target(df: pd.DataFrame, annual_growth_rate: float = 0.08) -> pd.DataFrame:
    """Creates the 'Future_Price_5Y' column (Regression Target)."""
    t = 5
    future_value_factor = (1 + annual_growth_rate) ** t
    df['Future_Price_5Y'] = df['Price_in_Lakhs'] * future_value_factor
    return df

# --- Classification Target ---
def create_good_investment_target(df: pd.DataFrame) -> pd.DataFrame:
    """Creates the 'Good_Investment' column (Classification Target) based on multi-factor score."""
    
    # Calculate City-level Price Per SqFt Median
    city_median_price = df.groupby('City')['Price_per_SqFt'].median().reset_index()
    city_median_price.rename(columns={'Price_per_SqFt': 'Median_Price_per_SqFt'}, inplace=True)
    df = df.merge(city_median_price, on='City', how='left')
    
    # Score 1: Value (1 if price is significantly cheaper than city median P/SqFt)
    df['Score_Value'] = np.where(df['Price_per_SqFt'] < (df['Median_Price_per_SqFt'] * 0.90), 1, 0)
    
    # Score 2: Infrastructure 
    df['Infra_Score'] = df['Public_Transport_Accessibility'].map({'Excellent': 3, 'Good': 2, 'Moderate': 1, 'Poor': 0}).fillna(0)
    df['Score_Infra'] = np.where(
        (df['Nearby_Schools'] >= 3) & 
        (df['Nearby_Hospitals'] >= 2) & 
        (df['Infra_Score'] >= 2), 
        1, 0
    )
    
    # Score 3: Property Demand/Quality 
    df['Score_Demand'] = np.where(
        (df['BHK'] >= 3) & 
        (df['Availability_Status'].isin(['Available', 'Ready to Move'])), 
        1, 0
    )
    
    # Final Classification: Good Investment if Total Score >= 2
    df['Total_Investment_Score'] = df['Score_Value'] + df['Score_Infra'] + df['Score_Demand']
    df['Good_Investment'] = np.where(df['Total_Investment_Score'] >= 2, 1, 0)
    
    df.drop(columns=['Median_Price_per_SqFt', 'Score_Value', 'Score_Infra', 'Score_Demand', 'Total_Investment_Score', 'Infra_Score'], inplace=True)
    return df

# Execute the functions
df = create_future_price_target(df)
df = create_good_investment_target(df)

print("Targets created. Sample:")
print(df[['Price_in_Lakhs', 'Future_Price_5Y', 'Good_Investment']].head())
```
