```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a style for better visualization
sns.set_style("whitegrid")
plt.figure(figsize=(12, 5))

# Price Distribution
plt.subplot(1, 2, 1)
sns.histplot(df['Price_in_Lakhs'], kde=True, bins=50)
plt.title('Distribution of Property Prices (in Lakhs)')
plt.xlabel('Price in Lakhs')

# Size Distribution
plt.subplot(1, 2, 2)
sns.histplot(df['Size_in_SqFt'], kde=True, bins=50)
plt.title('Distribution of Property Sizes (in SqFt)')
plt.xlabel('Size in SqFt')

plt.tight_layout()
plt.show()
```
